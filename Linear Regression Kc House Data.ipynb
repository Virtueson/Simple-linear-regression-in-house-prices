```python
import pandas as pd
import numpy as np
```

# Data Preprocessing

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

Function to return the year of the date column

```python
def slicing(x):
    x = x[0:4]
    return x
```

Function to show if the house has been renovated

```python
def renovated(x):
    if x>0:
        return 1
    else:
        return 0
```

Testing the function

```python
a = slicing(df.date[0])
a
```

# Applying function to a new column

```python
df['date_sold']= df.apply(lambda row : slicing(row['date']), axis = 1)
df.head()
```

```python
df.info(show_counts= True)
```

```python
df['bathrooms'] = df['bathrooms'].apply(np.floor)
df.head()
```

```python
df['renovated_times']= df.apply(lambda row : renovated(row['yr_renovated']), axis = 1)
df.head()
```

```python
inputs = df[['sqft_living','renovated_times','yr_built','sqft_lot']]
targets = df.price
```

```python
inputs.head()
```

```python
targets.head()
```

Linear Regression model application

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

Splitting the data into test and train data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,targets,test_size=0.2,random_state=10)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train,y_train)
```

The accuracy score of the model

```python
model.score(X_test,y_test)
```

The coefficient of each x value

```python
model.coef_
```
